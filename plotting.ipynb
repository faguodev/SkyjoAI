```python
import os
from collections import defaultdict

import matplotlib.pyplot as plt
import seaborn as sns
import json
import os
import numpy as np
sns.set()
```

```python
iteration = 0
episode_reward_mean = []
policy_reward_max = defaultdict(list)
policy_reward_mean = defaultdict(list)
while os.path.exists(f"logs1/result_iteration_{iteration}.json"):
    with open(f'logs1/result_iteration_{iteration}.json') as f:
        d = json.load(f)
        episode_reward_mean.append(d["env_runners"]["episode_reward_mean"])
        for policy, max_val in d["env_runners"]["policy_reward_max"].items():
            policy_reward_max[policy].append(max_val)
        for policy, max_val in d["env_runners"]["policy_reward_mean"].items():
            policy_reward_mean[policy].append(max_val)
    iteration += 5
```

```python
sns.lineplot(x=np.arange(0, iteration, 5), y=episode_reward_mean)
plt.title("Episode Mean Reward")
plt.xlabel("Iteration")
plt.show()
```

```python
for policy in policy_reward_max:
    plt.plot(np.arange(0, iteration, 5), policy_reward_max[policy], label=policy) #x=np.arange(0, iteration, 5), y=policy_reward_max[policy])

plt.legend()
plt.title("Policy Max Reward")
plt.xlabel("Iteration")
plt.show()
```

```python
for policy in policy_reward_mean:
    plt.plot(np.arange(0, iteration, 5), policy_reward_mean[policy], label=policy) #x=np.arange(0, iteration, 5), y=policy_reward_max[policy])

plt.rcParams["figure.figsize"] = (15, 10)
plt.legend()
plt.title("Policy Mean Reward")
plt.xlabel("Iteration")
plt.show()
```
